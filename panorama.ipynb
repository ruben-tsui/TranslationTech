```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.error import HTTPError
import re
```

```python
webpage = 'https://www.taiwan-panorama.com/Articles/Details?Guid=7269bd3e-9fca-4921-ba4c-70ee1cd780a9&CatId=5#lang-modal'

webpage = 'https://www.taiwan-panorama.com/Articles/Details?Guid=f3ef5956-780e-43a8-bcd6-9bcf7a2e4b5a&CatId=7'

webpage = 'https://www.taiwan-panorama.com/Articles/Details?Guid=011049d3-d320-45eb-b96d-ee53d5d6586a&CatId=7'
```

```python
try:
    html = urlopen(webpage)
except HTTPError as e:
    print(e)
except URLError as e:
    print('The server could not be found!')
else:
    print('It Worked!')
    bs = BeautifulSoup(html.read(), 'lxml')
```

### Chinese

```python
main_div_zh = bs.body.find('div', {'class': "row clearfix", 'id': "detail_Text"}).find('div', {'class': "col-xs-12"})
cnt = 0
for node in main_div_zh.find_all(re.compile(r"\bh[12]\b|\bp\b")):
    txt = node.get_text()
    if txt.strip() != '':
        print(f"[{node.name}] {txt}")
        cnt += 1
        print("-"*60)
print(f"cnt = {cnt}")
print("="*60)
```

### English and Japanese

```python
for part in bs.body.find_all('div', {'class': "col-lg-8 col-lg-offset-2"}):
    cnt = 0
    for node in part:
        if node.name in ['h2', 'p']:
            txt = node.get_text()
            if txt.strip() != '':
                print(f"[{node.name}] {txt}")
                cnt += 1
                print("-"*60)
    print(f"cnt = {cnt}")
    print("="*60)
```
